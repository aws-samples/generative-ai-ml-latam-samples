```python
#THIS CODE WAS COPY PASTED FROM ABOVE WITHOUT MODIFICATION !

import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d

# Create synthetic data mimicking the training progression
np.random.seed(42)  # For reproducibility

# Generate x-axis: training steps
steps = np.arange(0, 8001, 50)  # 0 to 8000 in steps of 50

# Generate synthetic average length data with increasing trend and noise
base_trend = 500 + 7500 * (1 - np.exp(-steps/4000))  # Exponential growth that plateaus
noise = np.random.normal(0, 500, size=len(steps))  # Random variations
smoothed_noise = gaussian_filter1d(noise, sigma=3)  # Smooth the noise

# Final synthetic data
avg_length = base_trend + smoothed_noise

# Generate synthetic variance data
variance = 1000 + steps/10  # Increasing variance over time
upper_bound = avg_length + variance
lower_bound = np.maximum(avg_length - variance, 0)  # Prevent negative values

# Smooth the curves for visualization
smooth_avg = gaussian_filter1d(avg_length, sigma=5)

# Create the plot
plt.figure(figsize=(12, 7))

# Plot shaded area for variance
plt.fill_between(steps, lower_bound, upper_bound, alpha=0.3, color='blue')

# Plot the average length
plt.plot(steps, smooth_avg, color='blue', linewidth=2)

# Add labels and title
plt.xlabel('Steps', fontsize=14)
plt.ylabel('Average length per response', fontsize=14)
plt.title('DeepSeek-R1-Zero average length per response during training', fontsize=16)

# Add grid
plt.grid(True, alpha=0.3)

# Set y-axis limit similar to the original figure
plt.ylim(0, 13000)

plt.tight_layout()
plt.savefig('deepseek_r1_zero_length_trend.png', dpi=300)
plt.show()

print("This visualization mimics the trend shown in Figure 3 of the DeepSeek-R1 paper.")
print("In reality, to create the actual figure, you would need:")
print("1. Access to the training logs from DeepSeek-R1-Zero's RL training")
print("2. Response length data collected at each training step")
```
